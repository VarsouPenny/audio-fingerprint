```python
import numpy as np 
import matplotlib.pyplot as plt 
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
  
X = ['100 songs','1000 songs','5800 songs',]
valid = [89.9,95.3,91.1]
invalid = [10.1, 4.7, 8.9]
  
X_axis = np.arange(len(X))
plt.figure(figsize=(9, 9))  
plt.bar(X_axis - 0.2, valid, 0.4, label = 'Valid')
plt.bar(X_axis + 0.2, invalid, 0.4, label = 'Invalid')
addlabels(X,valid)
addlabels(X,invalid)
plt.xticks(X_axis, X)

plt.xlabel("Databases")
plt.ylabel("Matcing Percentage")
plt.title("2 sec Matcing Percentage")
plt.legend()

plt.savefig('2_sec_matching.png')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
  
X = ['5 secs','10 secs','15 secs','20 secs']
valid = [11,20,19,21]
invalid = [39, 29, 31, 28]
  
X_axis = np.arange(len(X))
plt.figure(figsize=(9, 9))  
plt.bar(X_axis - 0.2, valid, 0.4, label = 'Valid')
plt.bar(X_axis + 0.2, invalid, 0.4, label = 'Invalid')
addlabels(X,valid)
addlabels(X,invalid)
plt.xticks(X_axis, X)

plt.xlabel("Queries")
plt.ylabel("Songs")
plt.title("Microphone maching")
plt.legend()

plt.savefig('mic_matching.png')
plt.show()
```
